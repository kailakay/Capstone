```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()

import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.express as px
import plotly.tools as tls

import math

%store -r years

%matplotlib inline
```

```python
emissions_df=pd.read_csv('Data/emissions_df.csv')
emissions_df.head()
```

```python
area_abbrev = list(emissions_df['Area Abbreviation'].unique())
```

```python
# keys = area list
# values = total product sum by year (53)
def total_yearly_product_count(dfName, iter_list):
    """
    Input dfName and list of country abbreviations to iterate through 
    to sum production of each country by year.
    
    Returns info as a dictionary
    """
    new_df = {}
    for a in iter_list:
        yearly_production = []
        for year in years: 
            years_sum = (dfName[year][dfName['Area Abbreviation'] == a].sum())
            yearly_production.append(years_sum * 1000) # The unit of measurement is by 1000 tonnes
        new_df[a] = yearly_production
    return new_df 
```

```python
data = total_yearly_product_count(emissions_df, area_abbrev)
```

```python
total_em_sum_df = pd.DataFrame.from_dict(data, orient='index', columns=years).reset_index().rename(
                                                            columns={'index': 'Area Abbreviation'})
total_em_sum_df.head()
```

## 2013 Emissions by Country

```python
fig = px.choropleth(total_em_sum_df, locations="Area Abbreviation",
                    color="Y2013",
                    color_continuous_scale="YlGnBu")
fig.write_image("Images/2013_world_emissions.png")
fig.write_html("Images/2013_world_emissions.html")
fig.show()
```

## 1961 Emissions by Country

```python
fig = px.choropleth(total_em_sum_df, locations="Area Abbreviation",
                    color="Y1961",
                    color_continuous_scale="YlGnBu")
fig.write_image("Images/1961_world_emissions.png")
fig.write_html("Images/1961_world_emissions.html")
fig.show()
```

```python
area = emissions_df['Area Abbreviation'].unique()
```

```python
# Annual Emissions by Country
plt.figure(figsize=(24,12))
for a in area:
    values = []
    for year in years:
        this_value = emissions_df[year][emissions_df['Area Abbreviation'] == a].sum()
        values.append(math.sqrt(this_value))
    plt.plot(values, label=a)
plt.xticks(np.arange(53), tuple(years), rotation=60)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=8, mode="expand", borderaxespad=0.)
plt.title('Annual Emissions by Country')
plt.show()
```

```python
abbrev_top = ['CHN', 'IND', 'USA', 'BRA', 'RUS']

plt.figure(figsize=(24,12))
for ar in abbrev_top:
    yearly_em = []
    for yr in years:
        yearly_em.append(emissions_df[yr][emissions_df['Area Abbreviation'] == ar].sum())
    plt.plot(yearly_em, label=ar)
plt.xticks(np.arange(53), tuple(years), rotation=60)
plt.legend(loc=2, ncol=2)
#plt.savefig('p.png')
plt.title('Emissions of China, India, USA, Brazil, and Russia')
plt.show()
```
